## Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating(rating):
    if rating>7.0:
        return "Excellent"
    elif rating>4.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Remark']=data['Rating'].apply(rating)
```

## Find Unique Values From Genre

```python
data['Genre']
```

```python
# Creating new list from Genre column of dataset
newli=[]
for i in data['Genre']:
    newli.append(i.split(','))
```

```python
newli
```

```python
# Creating 1D list from 2d list
newli_1D=[]
for i in newli:
    for a in i:
        newli_1D.append(a)
```

```python
newli_1D
```

```python
#Sorting unique values in different list

uniqli=[]
for i in newli_1D:
    if i not in uniqli:
        uniqli.append(i)
```
